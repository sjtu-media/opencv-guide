# Working with pytorch

```python
import cv2
import matplotlib.pyplot as plt
import torch

image_sjtu = cv2.imread("images/sjtu.jpeg")
image_sjtu = cv2.cvtColor(image_sjtu, cv2.COLOR_BGR2RGB)

# 1. to pytorch
tensor = torch.from_numpy(image_sjtu)

# 2. process (do some funny)
# pytorch[bs, ch, h, w] opencv (h, w, ch)
# 
tensor = tensor.permute([2, 0, 1]).unsqueeze(0)
# xxxxxxx

# 3. convert tensor to numpy, 2 image, 
image = tensor[0, :, :, :].permute([1, 2, 0]).numpy()


plt.imshow(image)
```
